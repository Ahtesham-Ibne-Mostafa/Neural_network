```python
import tensorflow as tf
import numpy as np
import pandas as pd
import math
```

```python
df = pd.read_csv('/content/Training.csv')
df.head()
```

```python
list(df.columns)
```

```python
df.shape
```

```python
null_values = df.isnull().sum()

print(list(null_values))
```

```python
df.describe()
```

```python
target='prognosis'
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

df[target] = encoder.fit_transform(df[target])
```

```python
labels=df[target]
label_mapping = {label: i for i, label in enumerate(np.unique(labels))}
num_classes = len(label_mapping)
labels_one_hot = tf.keras.utils.to_categorical([label_mapping[label] for label in labels], num_classes)
print (num_classes)
```

```python
# Define the training data

X_train= df.loc[:,df.columns != target] #features
y_train = labels_one_hot
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(128, activation='relu', input_shape=(None, 132)),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(num_classes, activation='softmax')
])

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Train the model

model.fit(X_train,y_train, epochs=10, batch_size=None)
```

```python
test_df=pd.read_csv('/content/Testing.csv')
```

```python
test_df[target] = encoder.fit_transform(test_df[target])
```

```python
test_labels=test_df[target]
test_label_mapping = {label: i for i, label in enumerate(np.unique(test_labels))}
num_classes = len(test_label_mapping)
test_labels_one_hot = tf.keras.utils.to_categorical([test_label_mapping[label] for label in test_labels], num_classes)
```

```python
X_test= test_df.loc[:,test_df.columns != target] #features
y_test= test_labels_one_hot
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(y_test,y_pred)) # model prediction
```
